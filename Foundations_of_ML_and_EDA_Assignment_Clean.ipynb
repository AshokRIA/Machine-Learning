### Question 1
**Difference between AI, ML, DL, and Data Science.**

**Answer:**

- **AI (Artificial Intelligence):** Broad field creating machines that simulate human intelligence.
- **ML (Machine Learning):** Subset of AI where systems learn from data without explicit programming.
- **DL (Deep Learning):** Subset of ML using multi-layered neural networks for complex tasks like image or speech recognition.
- **Data Science:** Involves data collection, analysis, visualization, and modeling — using AI/ML to extract insights.

### Question 2
**Explain overfitting and underfitting in ML. How can you detect and prevent them?**

**Answer:**

- **Overfitting:** Model learns noise and performs well on training but poorly on test data.
- **Underfitting:** Model is too simple and fails to learn patterns.
- **Detection:** Compare training vs. validation accuracy.
- **Prevention:** Use cross-validation, regularization (L1/L2), and simpler models or more data.

### Question 3
**How would you handle missing values in a dataset? Explain at least three methods.**

**Answer:**

- **Deletion:** Remove rows/columns with missing values (use cautiously).
- **Imputation:** Replace with mean/median/mode based on distribution.
- **Predictive Modeling:** Use regression or ML models to predict missing values based on other features.

### Question 4
**What is an imbalanced dataset? Describe two techniques to handle it.**

**Answer:**

- **Imbalanced dataset:** When classes are unequally represented (e.g., 90% negative, 10% positive).
- **SMOTE:** Synthetic Minority Oversampling Technique creates synthetic samples for minority class.
- **Class Weights:** Adjust model’s loss function to penalize misclassification of minority class.

### Question 5
**Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.**

**Answer:**

- **Importance:** Many algorithms (KNN, SVM, gradient descent) rely on distance or gradients; scaling ensures fair contribution of features.
- **Min-Max Scaling:** Scales features to a fixed range [0,1].
- **Standardization:** Centers data around mean 0, variance 1. Better for data with outliers.

### Question 6
**Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?**

**Answer:**

- **Label Encoding:** Converts categories into numeric labels (0,1,2,...). Used for **ordinal** data where order matters.
- **One-Hot Encoding:** Creates binary columns for each category. Used for **nominal** data where order doesn’t matter.
